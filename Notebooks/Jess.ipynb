```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import citipy as citipy
import numpy as np
```

```python
filepath= os.path.join('..', 'Data','clean_data','clean_data.csv')
filepath
```

```python
df = pd.read_csv(filepath)
df.head(45)
```

```python
state_count = df['state'].value_counts()
unique_states = df["state"].unique()
unique_per_capita = df[['state','incidents per 10k']]
state_count = pd.DataFrame(state_count)
state_count = state_count.reset_index()
state_count = state_count.rename(columns = {'state':'Number of incidents'})
state_count = state_count.rename(columns = {'index':'state'})
state_count
```

```python
dataset = unique_per_capita.merge(state_count, how = 'left')
dataset = dataset.drop_duplicates('state')
dataset = dataset.rename(columns = {'state':'State'})
dataset = dataset.rename(columns = {'incidents per 10k':'Incident Per Capita'})
dataset
```

```python
organized_df = pd.DataFrame(dataset).sort_values(by='Number of incidents', ascending=False)
organized_df = organized_df.reset_index()
organized_df = organized_df.drop(['index'],axis = 1)
topten = organized_df[0:10]
topten
```

```python
x_axis = topten['State']
plt.bar(x_axis, topten['Number of incidents'], color="purple", align="center")
plt.xticks(rotation = 75)
axes2 = plt.twinx()
axes2.plot(x_axis,topten['Incident Per Capita'], color = 'pink',linewidth = 2, marker = 'o')
```

```python
organized_pc = pd.DataFrame(dataset).sort_values(by='Incident Per Capita', ascending=False)
organized_pc = organized_pc.reset_index()
organized_df = organized_pc.drop(['index'],axis = 1)
topten_pc = organized_df[0:10]
topten_pc
```
