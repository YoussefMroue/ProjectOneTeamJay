```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
file = os.path.join("..","Data","clean_data","clean_data.csv")
raw = pd.DataFrame(pd.read_csv(file))

main = pd.DataFrame(raw[['incident_id','date','state','latitude','longitude','incident_characteristics']])
main.fillna('BLANK')
main['Type'] = ""
test = main.head(20)
```

```python
incident_type_dict= {
    "Gang involvement":0,
    "Drug involvement":0,
    "Domestic Violence":0,
    "Armed robbery":0,
    "Officer Involved Incident":0,
    "Home Invasion":0,
    "Car-jacking":0,
    "Negligent Discharge":0,
    "suicide":0,
    "Child Involved":0,
    "Pistol-whipping":0,
    "Defensive Use":0,
    "Accidental Shooting":0,
    "No Injuries":0,
    "Home Invasion":0,
    "Murder":0,
    "Road rage":0,
    "Animal shot":0,
    "Drive-by":0,
    "Police Targeted":0,
    "School Incident":0,
    "Hate crime":0,
    "Terrorism Involvement":0,
    "Assault weapon":0,
    "Mass Shooting":0,
    "Accidental Shooting":0,
    "Unknown":0}
incident_type = ["Gang involvement",
    "Drug involvement",
    "Domestic Violence",
    "Armed robbery",
    "Officer Involved Incident",
    "Home Invasion",
    "Car-jacking",
    "Negligent Discharge",
    "suicide",
    "Child Involved",
    "Pistol-whipping",
    "Defensive Use",
    "Accidental Shooting",
    "No Injuries",
    "Home Invasion",
    "Murder",
    "Road rage",
    "Animal shot",
    "Drive-by",
    "Police Targeted",
    "School Incident",
    "Hate crime",
    "Terrorism Involvement",
    "Assault weapon",
    "Mass Shooting",
    "Accidental Shooting"]
```

```python
type_inc = []
for entry in main['incident_characteristics']:
    temp = []
    for i in incident_type:
        if i in str(entry):
            if i in str(temp):
                continue
            else:
                temp.append(i)
                incident_type_dict[i]+=1
    if temp == []:
        temp = ['Blank']
        incident_type_dict['Unknown']+=1

    type_inc.append(temp)
incident_type_dict
```

```python
main['Type'] = type_inc
main.head()
```

```python
states = main['state'].unique()
type_state_count = pd.DataFrame(states)
type_state_count = type_state_count.rename(columns = {0:'State'})
type_state_count = type_state_count.set_index('State')
for i in incident_type:
    type_state_count[i]=0


```

```python
for index,row in main.iterrows():
    State = row[2]
    current = row[6]
    for inc in incident_type:
        if inc in current:
            type_state_count.loc[State,inc]+=1
            
```

```python
years = ['2013','2014','2015','2016','2017','2018']
type_year_count = pd.DataFrame(years)
type_year_count = type_year_count.rename(columns = {0:'Year'})
type_year_count = type_year_count.set_index('Year')
for i in incident_type:
    type_year_count[i]=0
```

```python
for index,row in main.iterrows():
    year = row[1][:4]
    current = row[6]
    for inc in incident_type:
        if inc in current:
            type_year_count.loc[year,inc]+=1
```
