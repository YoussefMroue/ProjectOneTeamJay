```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import operator
import gmaps

from google_maps_key import gkey
```

```python
file = os.path.join("..","Data","clean_data","clean_data.csv")
raw = pd.DataFrame(pd.read_csv(file))

main = pd.DataFrame(raw[['incident_id','date','state','latitude','longitude','incident_characteristics']])
main.fillna('BLANK')
main['Type'] = ""
main.head()
```

```python
incident_type_dict= {
    "Gang involvement":0,
    "Drug involvement":0,
    "Domestic Violence":0,
    "Armed robbery":0,
    "Officer Involved Incident":0,
    "Home Invasion":0,
    "Car-jacking":0,
    "Negligent Discharge":0,
    "suicide":0,
    "Child Involved":0,
    "Pistol-whipping":0,
    "Defensive Use":0,
    "Accidental Shooting":0,
    "No Injuries":0,
    "Home Invasion":0,
    "Murder":0,
    "Road rage":0,
    "Animal shot":0,
    "Drive-by":0,
    "Police Targeted":0,
    "School Incident":0,
    "Hate crime":0,
    "Terrorism Involvement":0,
    "Assault weapon":0,
    "Mass Shooting":0,
    "Accidental Shooting":0,
    "Unknown":0}
incident_type = ["Gang involvement",
    "Drug involvement",
    "Domestic Violence",
    "Armed robbery",
    "Officer Involved Incident",
    "Home Invasion",
    "Car-jacking",
    "Negligent Discharge",
    "suicide",
    "Child Involved",
    "Pistol-whipping",
    "Defensive Use",
    "Accidental Shooting",
    "No Injuries",
    "Home Invasion",
    "Murder",
    "Road rage",
    "Animal shot",
    "Drive-by",
    "Police Targeted",
    "School Incident",
    "Hate crime",
    "Terrorism Involvement",
    "Assault weapon",
    "Mass Shooting",
    "Accidental Shooting"]
```

```python
type_inc = []
for entry in main['incident_characteristics']:
    temp = []
    for i in incident_type:
        if i in str(entry):
            if i in str(temp):
                continue
            else:
                temp.append(i)
                incident_type_dict[i]+=1
    if temp == []:
        temp = ['Blank']
        incident_type_dict['Unknown']+=1

    type_inc.append(temp)
sorted_incident_type_dict = sorted(incident_type_dict.items(), key = operator.itemgetter(1))
sorted_incident_type_dict
```

```python
main['Type'] = type_inc
main.head()
```

```python
states = main['state'].unique()
type_state_count = pd.DataFrame(states)
type_state_count = type_state_count.rename(columns = {0:'State'})
type_state_count = type_state_count.set_index('State')
for i in incident_type:
    type_state_count[i]=0


```

```python
for index,row in main.iterrows():
    State = row[2]
    current = row[6]
    for inc in incident_type:
        if inc in current:
            type_state_count.loc[State,inc]+=1
            
```

```python
type_state_count.head()
```

```python
output_file=os.path.join('output','incident_type.csv')
type_state_count.to_csv(output_file,index=True,header=True)
```

```python
years = ['2013','2014','2015','2016','2017','2018']
type_year_count = pd.DataFrame(years)
type_year_count = type_year_count.rename(columns = {0:'Year'})
type_year_count = type_year_count.set_index('Year')
for i in incident_type:
    type_year_count[i]=0
```

```python
for index,row in main.iterrows():
    year = row[1][:4]
    current = row[6]
    for inc in incident_type:
        if inc in current:
            type_year_count.loc[year,inc]+=1
```

```python
gmaps.configure(api_key = gkey)
maps_df = main[['incident_id','latitude','longitude','Type']]
maps_df.head()
```

```python
maps_df = maps_df.fillna('drop')
for index, row in maps_df.iterrows():
    if row[1] == 'drop':
        maps_df = maps_df.drop([index])
maps_df.head()
```

```python
output_file=os.path.join('output','incident_type_gmap.csv')
maps_df.to_csv(output_file,index=False,header=True)
```

```python
top_five_inc = ['suicide','Officer Involved Incident','Armed robbery','No Injuries','Drug involvement']

suicides_lat = []
suicides_lon = []

officers_lat = []
officers_lon = []

armed_lat = []
armed_lon = []

injuries_lat = []
injuries_lon = []

drug_lat = []
drug_lon = []

for index, row in  maps_df.iterrows():
    if  top_five_inc[0] in row[3]:
        
        suicides_lat.append(row[2])
        suicides_lon.append(row[1])
        
    elif top_five_inc[1] in row[3]:
        
        officers_lat.append(row[2])
        officers_lon.append(row[1])
        
    elif top_five_inc[2] in row[3]:
        
        armed_lat.append(row[2])
        armed_lon.append(row[1])
        
    elif top_five_inc[3] in row[3]:
        
        injuries_lat.append(row[2])
        injuries_lon.append(row[1])
        
    elif top_five_inc[4] in row[3]:
        drug_lat.append(row[2])
        drug_lon.append(row[1])
        
        
            

        
```

```python
drugs = pd.DataFrame(drug_lat,drug_lon)
drugs = drugs.reset_index()

suicides = pd.DataFrame(suicides_lat,suicides_lon)
suicides = suicides.reset_index()

officers = pd.DataFrame(officers_lat,officers_lon)
officers = officers.reset_index()

injuries = pd.DataFrame(injuries_lat,injuries_lon)
injuries = injuries.reset_index()

armed = pd.DataFrame(armed_lat,armed_lon)
armed = armed.reset_index()
```

```python
armed_layer = gmaps.symbol_layer(armed, fill_color = 'rgba(255,248,117,1)',
                                stroke_color = 'rgba(255,248,117,1)',scale = 2
                                )
drugs_layer = gmaps.symbol_layer(drugs, fill_color = 'rgba(255,138,194,1)',
                                stroke_color = 'rgba(255,138,194,1)',scale = 2
                                )
suicides_layer = gmaps.symbol_layer(suicides, fill_color = 'rgba(230,0,0,1)',
                                stroke_color = 'rgba(230,0,0,1)',scale = 2
                                )
officers_layer = gmaps.symbol_layer(officers, fill_color = 'rgba(0,15,230,1)',
                                stroke_color = 'rgba(0,15,230,1)',scale = 2
                                )
injuries_layer = gmaps.symbol_layer(injuries, fill_color = 'rgba(255,255,255,1)',
                                stroke_color = 'rgba(255,255,255,1)',scale = 2
                                )
fig = gmaps.figure()
fig.add_layer(drugs_layer)
# fig.add_layer(armed_layer)
# fig.add_layer(suicides_layer)
# fig.add_layer(officers_layer)
# fig.add_layer(injuries_layer)
fig
```
