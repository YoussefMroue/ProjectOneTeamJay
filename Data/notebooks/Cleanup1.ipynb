```python
import os 
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
file = os.path.join("..","raw_data","gun-violence-data_01-2013_03-2018.csv")
raw_df = pd.read_csv(file)
raw_df = pd.DataFrame(raw_df)
raw_df.head()
```

```python
all_df =pd.DataFrame(raw_df[['incident_id','date','state','city_or_county','latitude','longitude','location_description','n_killed','n_injured','n_guns_involved','gun_stolen','gun_type','participant_age','participant_age_group','participant_gender','participant_status','participant_type','incident_characteristics','source_url']])
all_df.head()



```

```python
state_count = pd.DataFrame(all_df['state'].value_counts())

state_count= state_count.rename(columns = {'state':'Incident Count'})
state_count.head()
```

```python
state_gb = all_df.groupby(['state', 'city_or_county'])
state_gb = pd.DataFrame(state_gb['incident_id'].count())
state_gb=state_gb.reset_index()
state_gb
```
