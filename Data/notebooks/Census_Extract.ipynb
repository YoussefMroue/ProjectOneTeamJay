```python
## Import dependices, 

from census import Census #python wrapper for for census API
from us import states
import pandas as pd
from census_api_key import api_key
import os

#provide api key and year of desired data
c=Census(api_key, year=2017)
```

```python
## using the acs5 function set to pull our data from the 5 year american consumer survey
## using wrapper function to request api 

census_data=c.acs5.get(('NAME','B01003_001E'), {'for':'state:*'})
census_data
```

```python
## create a dataframe of data pulled from API call

census_df=pd.DataFrame(census_data)
census_df
```

```python
census_output=os.path.join('..','raw_data','census_data.csv')
census_df.to_csv(census_output,header=True)
```
